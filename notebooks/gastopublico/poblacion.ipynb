```python
import pandas as pd
import numpy as np
```

```python
poblacion = pd.read_excel("../../data/external/reporte_estadist.xlsx", dtype = str)
```

```python
poblacion.columns = list(poblacion.columns[:-2]) + [2007, 2017]
```

```python
poblacion
```

```python
poblacion = poblacion.drop([0])
```

```python
poblacion[2007] = poblacion[2007].fillna(1000).dropna().apply(int)
```

```python
poblacion[2017] = poblacion[2017].dropna().apply(int)
```

```python
poblacion
```

```python
Poblacionperregion = (poblacion
                      .groupby(['Departamento','Provincia'])
                      .apply(lambda y: pd.Series(
                          {2007: y[2007].sum(), 
                           2017: y[2017].sum(), 
                           'Ubigeo': y['Ubigeo'].iloc[0][0:4]
                          }
                      )
                            )
                     )
                                                      
```

```python
Poblacionperregion
```

```python
g_rate = (Poblacionperregion[2017]/(Poblacionperregion[2007]))**(1/10)

for year in range(2008, 2022):
    Poblacionperregion[year] = Poblacionperregion[year-1]* g_rate
```

```python
Poblacionperregion
```

```python
Poblacionperregion.to_pickle("../../data/raw/poblacion-provincia-periodo.pickle")
```
